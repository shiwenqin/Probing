```python
def add_row(df,model,task,performance):
    if model.endswith('base'):
        model_type = 'base'
    else:
        model_type = 'large'
    for index, acc in enumerate(performance):
        df.loc[len(df)] = [index,model,task,acc,model_type]
    return df

df = add_row(df,'Bert-base','POS',bert_pos_base)
df = add_row(df,'Albert-base','POS',albert_pos_base)
df = add_row(df,'Bert-large','POS',bert_pos_large)
df = add_row(df,'Albert-large','POS',albert_pos_large)

df = add_row(df,'GPT2-base','POS',gpt2_pos_base)

df = add_row(df,'Bert-base','RE',bert_base_re)
df = add_row(df,'Albert-base','RE',albert_base_re)
df = add_row(df,'Bert-large','RE',bert_large_re)
df = add_row(df,'Albert-large','RE',albert_large_re)
```

```python
sns.set_theme(style="whitegrid")
sns.lineplot(x="Layer", y="Accuracy", hue="Model", data=df[(df['Type']=='base') & (df['Task']=='POS')])
plt.title('POS Tagging Accuracy for Probing Bert (base)')
```

```python
sns.set_theme(style="whitegrid")
sns.lineplot(x="Layer", y="Accuracy", hue="Model", data=df[(df['Type']=='large') & (df['Task']=='POS')])
plt.title('POS Tagging Accuracy for Probing Bert (large)')
```
